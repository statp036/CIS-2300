```python
from random import randint

#character class for our hero
class Character:
    inventory = [[0,0,0,0,0],["Potion", "Fire Ring", "Earth Ring", "Water Ring", "Bless Ring"]] #player inventory
    name = "Protaginst"
    #stats of player
    def __init__(self):
        self.mhp = 620
        self.hp = 620
        self.strength = 50
        self.magic = 50
        self.luck = 10
        self.defense = 50
        self.alive = True
    
    #change name of character
    def changename(self, name):
        self.name = name
    
    #add item to an inventory
    def addInventory(self, num, amount):
        self.inventory[0][num] += amount
    
    #print stats of user
    def printstats(self):
        print(f"HP: {self.hp}")
        print(f"Strength: {self.strength}")
        print(f"Magic: {self.magic}")
        print(f"Defense: {self.defense}")
        print(f"Luck: {self.luck}")
    
    # function to equip item
    #potion use 
    def equip(self, num, consume):
        if(num-1 == 0 and self.inventory[0][0] > 0 and consume == 0):
            if(self.hp < self.mhp - 500):
                self.hp += 500
            else:
                self.hp = 620
            self.inventory[0][0] += -1
        #equip items and check if item is available to use
        elif(num == 1 and self.inventory[0][1] > 0):
            self.strength = self.strength * 1.5
            self.inventory[0][1] += -1
        elif(num == 2 and self.inventory[0][2] > 0):
            self.defense = self.defense * 1.5
            self.inventory[0][2] += -1  
        elif(num == 3 and self.inventory[0][3] > 0):
            self.magic = self.magic * 1.5
            self.inventory[0][3] += -1
        elif(num == 4 and self.inventory[0][4] > 0):
            self.luck = self.luck + 10
            self.inventory[0][4] += -1
        else:
            print("No items to equip")
            
#stat modifier for human race
class Human(Character):
    def __init__(self):
        super().__init__()
    #modify stats according to race
    def stats(self):
        self.strength = int((1+ randint(10,20)/100) * self.strength)
        self.magic = int((randint(60,80)/100) * self.magic)
        self.defense = int(randint(80,90)/100 * self.defense) 
        
#stat modifier for elf race
class Elf(Character):
    def __init__(self):
        super().__init__()
    #modify stats according to race
    def stats(self):
        self.strength = int((randint(60,80)/100) * self.strength)
        self.magic = int((1 + randint(40,50)/100) * self.magic)
        self.defense = int(randint(60,80)/100 * self.defense)
          
#mage class
class mage(Elf):
    #base power of moves
    __fire = 60
    __wind = 40
    __light = 100
    def __init__(self):
        super().__init__()
        self.movelist = ["Sage's Fire", "Blizzard", "Astral Front"] #list of moves
    #moves for the mage class and calculate damage
    def move(self,selection):
        if(selection == 1):
            x = self.magic + self.__fire 
            print("Sage's Fire!!")
        elif(selection == 2):
            x = self.magic * 2 + self.__wind
            print("Blizzard!!")
        else:
            x = self.magic * 3 + self.__light
            print("Astral Front!!")
        return x

#mage knight class     
class mageKnight(Elf):
    #base power of moves
    __fireb = 30
    __healb = 20
    __mysticb = 70
    def __init__(self):
        super().__init__()
        self.movelist = ["Blazing Edge", "Siphon Slash", "Arcanic Blade"] #list of moves
    #moves for the mage knight class and calculate damage
    def move(self,selection):
        if(selection == 1):
            x = self.strength * 0.7 + self.magic + self.__fireb
            print("Blazing Edge!!")
        elif(selection == 2):
            x = self.strength * 0.3 + self.magic + self.__healb
            print("Siphon Slash!!")
            if(self.hp + x <= 620):
                self.hp = x + self.hp
        else:
            x = self.strength * 0.5 + self.magic * 2.5 + __mysticb
            print("ARCANIC BLADE!!")
        return x

#archer class
class archer(Human):
    #base power of moves
    __parrow = 80
    __deye = 120
    def __init__(self):
        super().__init__()
        self.movelist = ["Piercing Arrow", "Deadeye"] #list of moves
    #moves for the archer class and calculate damage
    def move(self, selection):
        if(selection == 1):
            x = self.strength + self.__parrow * 1.5
            print("PIERCING ARROW!!")
        else:
            x = self.strength + self.__deye * 2
            print("DEADEYE!!")
        return x          
    
#palidan class
class palidan(Human): 
    #base power of moves
    __dslash = 60
    __sdance = 100
    def __init__(self):
        super().__init__()
        self.movelist = ["Double Slash", "Sword Dance"] #calculate damage
    #moves for the palidan class and calculate damage
    def move(self, selection):
        if(selection == 1):
            x = self.strength + self.__dslash * 2
            print("DOUBLE SLASH!!")
        else:
            x = self.strength * 3 + self.__sdance
            print("SWORD DANCE!!")
        return x    

#set up imp enemy
class imp():
    #base power of moves
    __bite = 20
    __scratch = 30
    #stats for imp
    def __init__(self):
        self.hp = 150
        self.strength = 30
        self.defense = 20 
        self.luck = 0
        self.alive = True
        self.name = "Imp"
    #moves imp use
    def move(self,selection):
        if(selection == 1):
            x = self.strength + self.__bite
            print("Imp used bite")
        else:
            x = self.strength + self.__scratch
            print("Imp used scratch")
        return x
    
#set up general miniboss
class general():
    #base power of moves
    __cleave = 70
    __charge = 90
    #stats for general
    def __init__(self):
        self.hp = 600
        self.strength = 90
        self.defense = 90
        self.luck = 0
        self.alive = True
        self.name = "General"
    #moves general use
    def move(self,selection):
        if(selection == 1):
            x = self.strength * 1.1 + self.__cleave
            print("General used heavy cleaver")
        else:
            x = self.strength + self.__charge
            print("General used unwavering charge")
        return x   
    
#set up legion enemy   
class legion():
    #base power of moves
    __thrust = 40
    __sweep = 50
    #stats for legion
    def __init__(self):
        self.hp = 400
        self.strength = 50
        self.defense = 75
        self.luck = 0
        self.alive = True
        self.name = "Legion"
    #moves legion use
    def move(self,selection):
        if(selection == 1):
            x = self.strength + self.__thrust
            print("Legion used spear thrust")
        else:
            x = self.strength + self.__sweep
            print("Legion used wide sweep")
        return x
    
#set up dark lord boss
class darklord():
    #base power of moves
    __rupt = 80
    __chaos = 100
    #stats for dark lord
    def __init__(self):
        self.hp = 1500
        self.strength = 100
        self.defense = 70
        self.luck = 0
        self.alive = True
        self.name = "Dark Lord"
    #moves dark lord used
    def move(self,selection):
        if(selection == 1):
            x = self.strength + self.__rupt
            print("Dark Lord used quaking rupture")
        else:
            x = self.strength + self.__chaos
            print("Dark Lord used mass chaos")
        return x

#function to perform combat
def combat(player, enemy):
    #ends when either dies
    while(player.alive == True or enemy.alive == True):
        print("1.Battle 2.Item")
        action = int(input(f"What will {player.name} do (integer)? "))
        #see if player can use items 
        if(action != 1 and player.inventory[0][0] != 0):
            print(f"1. {player.inventory[0][0]} x {player.inventory[1][0]}")
            select = int(input("Choose a item (integer) "))
            player.equip(select,0)
        else:
            #print list of moves
            for i in range(len(player.movelist)):
                print(f"{i+1}. {player.movelist[i]}")
            select = int(input("Choose a move (integer)")) #player select move
            playermove = player.move(select)
            enemy.hp = int(enemy.hp - (playermove - enemy.defense)) #calculate damage
            print(f"{player.name} inflict {int(playermove - enemy.defense)} damage")  #tell hoe much damage is inflicted
        print(f"{enemy.name}: {enemy.hp} hp") #print hp of enemy 
        #end combat if enemy is defeated
        if(enemy.hp <= 0):
            enemy.alive = False
            print(f"{enemy.name} has been defeated")
            break
        #enemy choose a move
        enemymove = randint(1,2)
        rng = randint(1,100)  
        enemymove = enemy.move(enemymove)
        #see if player didn't dodged
        if(rng > player.luck):
            if(enemymove - player.defense > 0):
                player.hp = int(player.hp - (enemymove - player.defense))  #player hp reduced
                print(f"{enemy.name} has inflicted {int(enemymove - player.defense)}") #calculate damage on player
            else:
                print("You took no damage")
        else:
            print(f"{enemy.name} missed")
        #print hp of player
        print(f"{player.name}: {player.hp} hp")
        #check if player is defeated
        if(player.hp <= 0):
            player.alive = False
            print(f"You have been defeated")
            break

#ask player if they want to equip an item        
def intermission(player):
    temp = True
    while(temp == True): #end if player no longer want to equip items
        answer = str(input("Do you want to equip an item (Y/N)? "))
        if(answer == "Y" or answer == "y"):
            #print items
            for i in range(4):
                print(f"{i+1}. {player.inventory[0][i+1]} x {player.inventory[1][i+1]}")
            answer = int(input("Which item do you want to equip (integer)? "))
            #equip item
            if(player.inventory[0][answer] > 0):
                player.equip(answer,1)
            else:
                print("Please choose another item")
        else:
            print("Let's continue onward")
            temp = False
    
#create map
def create_map():
    pmap = [["E","*","*","*"],["*","S","*","R"],["*","*","G","*"],["*","*","*","C"]]
    for x in range(4):
        print(x,":",pmap[x][0],pmap[x][1],pmap[x][2],pmap[x][3])

#tutourial section
def entrance(player):
    enemy = imp() #create imp
    print("(You spawned at the entrance of the capital)")
    print("Resident1: Adventuuerer! It's finally good to see you. This is the capital and the dark lord is residing in the castle")
    print("His minions roam the place from the castle all the way to the town square please helps us")
    print("Child: An IMP has invaded please help us")
    #ask user if they are a resident
    choice = str(input(("Imp: You adventuerer. I've never seen you before. Are you a resident? (Y/N) ")))
    if(choice == "Y" or choice == "y"):
        print("Eh I don't care. You piss me off")
    else:
        print("DIE!!!")
    combat(player, enemy) #engage combat
    #check if player is alive
    if(player.alive == True):
        print("Reisdent1: Thank you traveler please don't waste your time go save the capital.") #end of tutourial
    else:
        print("Game Over")
        
#place to do quests
def square(player):
    boolquests = True
    quests= [["1. Clean the sewer", "2. Answer the seer riddle", "3. Slay the general"],[False,False,False]]
    print("You have arrived at the town square. It's empty but you see a quest board")
    while(boolquests == True and player.alive == True):
        choice = input("Do you want to do a quest (Y/N)? ")
        if(choice == "Y" or choice == "y"):
            for i in range(3):
                print(f"{quests[0][i]}")
            questsC = int(input("Which quest do you want to do (integer)? "))
            #sewer quest
            if(questsC == 1 and quests[1][questsC-1] == False):
                quests[1][questsC-1] = True
                print("You entered the sewer")
                print("Despite how dingy and murky it is you preveiled")
                #reward player
                print("Bless Ring acquired")
                player.addInventory(4,1)  #add bless ring to inventory
                intermission(player)    #open menu to equip item
            #riddle quest
            elif(questsC == 2 and quests[1][questsC-1] == False):
                quests[1][questsC-1] = True
                #ask riddle
                print("Seer: Answer me correctly and I shall bestow you this earth ring")
                answer = int(input("Seer: Now riddle me this, What is 25690*9087*0? "))
                #reward player
                if(answer == 0):  #check if user answered correctly
                    print("You are wise please take this")
                    print("Earth Ring acquired")
                    player.addInventory(2,1)   #add earth ring to inventory
                    intermission(player)      #open menu to reward player
                else:
                    print("You didn't answer my riddle correctly. Don't come back.")
            #slaying general quest
            elif(questsC == 3 and quests[1][questsC-1] == False):
                enemy = general()
                print("You went to fight the general")
                print("General: What? How dare you approach me like that ")
                combat(player, enemy) #engage combat
                #check if general has been defeated and reaward player
                if(enemy.alive == False):
                    print("General: I have been defeated.")
                    print("Fire Ring acquired")
                    player.addInventory(1,1)   #add fire ring to inventory
                    intermission(player)      #allow player to equip an item
                else: 
                    print("Gameover")
                    return 0;
                quests[1][questsC-1] = True
            #ask player to change or ignore quests
            else:
                print("You did this quests. Please select a new quest or leave.")
        #leave the town square
        else:
            boolquests = False
    print("You left the square ")

#fight castle guard
def gate(player):
    print("You have arrived at the castle's gate")
    enemy = legion() #create guard
    print("Legion: None shall pass this gate. Leave now or die")
    combat(player, enemy)
    #reward if guard is defeated
    if(enemy.alive == False):
        print("Water Ring acquired")
        player.addInventory(3,1)
    else:
        print("Game over")
    print("You have move towards the throne room")

#fight dark lord
def throne(player):
    print("After a long trek, you have arrived at the throne room")
    boss = darklord() #create boss
    print("Dark Lord: An adventuerer. How unexpected. I will not let you end me!")
    combat(player, boss)
    #check if dark lord is defeated
    if(boss.alive == False):
        print("Congrats you saved the capital")
    else:
        print("Gameover")

def main():
    #introduction
    print("Welcome to my rpg. Sadly you were reincarnated into my fantasy world where a dark lord has taken over the capital.")
    print("It is your job to defeat him. I will give you everything you need to aide you in battle")
    print("You can choose only two of these races: 1.Elf  2.Human ")
    #choosing race
    race_chosen = int(input("Please choose a race: 1.Elf 2.Human (integer) "))
    if(race_chosen == 1):
        class_chosen = int(input("Choose one of these two classes: 1.Mage 2.Mage Knight (integer) "))
        #choosing class
        if(class_chosen == 1):
            char1 = mage()
        else:
            char1 = mageKnight()
    else:
        #choosing class
        class_chosen = int(input("Choose one of these two classes: 1.Archer 2.Palidan (integer) "))
        if(class_chosen == 1):
            char1 = archer()
        else:
            char1 = palidan()
    name = str(input("Excuse my manners but what is your name? ")) #ask for name
    #give 10 potions
    print("I have already gave you 10 potions. Here is a map. Now go save the capital!") 
    char1.addInventory(0,10)
    create_map()  #create map
    char1.changename(name) #set name
    char1.stats()          #set stats
    char1.printstats()     #print stats
    entrance(char1)
    #game continue if player is alive
    if(char1.alive == True):
        square(char1)   #player go to square
    if(char1.alive == True):
        gate(char1)     #player go to castle gate
    if(char1.alive == True):   #prepare user to fight dark lord
        intermission(char1)
        char1.printstats()
        throne(char1)  
main()
```
